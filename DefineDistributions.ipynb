## Uniform Distribution

```python
# Practice generating a uniform distribution 
UniformDist = np.random.uniform(0.005, 20, size=20000)

# Plot the histogram of the bimodal distribution
plt.hist(UniformDist, bins=100, density=False, alpha=0.6, color='g')
plt.title('Uniform Concentrations Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("Max Concentration;",UniformDist.max().round(2), 'mM')
print("Min Concentration:",UniformDist.min().round(4), 'mM')
print("Mean Concentration:",UniformDist.mean().round(2), 'mM')
```

## Low Concentration Distribution

```python
# Practice generating a uniform distribution 
Dist1 = np.random.uniform(0.005, 0.05, size=10000)
Dist2 = np.random.uniform(0.005, 0.1, size=10000)
Dist3 = np.random.uniform(0.005, 0.2, size=10000)
LowConcDist = np.concatenate((Dist1, Dist2, Dist3))

# Plot the histogram of the bimodal distribution
plt.hist(LowConcDist, bins=100, density=False, alpha=0.6, color='g')
plt.title('Low Concentrations Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("Max Concentration;",LowConcDist.max().round(2), 'mM')
print("Min Concentration:",LowConcDist.min().round(4), 'mM')
print("Mean Concentration:",LowConcDist.mean().round(2), 'mM')



# Sort the samples to find the highest and lowest values
sorted_samples = np.sort(LowConcDist)
# Get the 50 lowest values
lowest_50 = sorted_samples[:100]
# Get the 50 highest values
highest_50 = sorted_samples[-100:]
# Print the 50 lowest values
print("50 Lowest Concentrations:", np.round(lowest_50, 4))
# Print the 50 highest values
print("50 Highest Concentrations:", np.round(highest_50, 3))
```

## Interference Signal Distribution

```python
# Practice generating a combined uniform/gamma distribution for interferences, similar to mimicking the overall distribution seen in tissues

InterferenceRange = 150*np.random.gamma(0.2, 0.02, size = 20000)
#InterferenceRange = np.concatenate((Dist1)) 
for i in np.arange(20000):
    InterferenceRange[i] += np.random.uniform(0.005, 0.5) + np.random.choice([0, np.random.uniform(0.005, 0.02)])

# Plot the histogram of the bimodal distribution
plt.hist(InterferenceRange, bins=100, density=False, alpha=0.6, color='g')
plt.title('Interference Range Concentrations Distribution')
plt.xlabel('Concentration (mM)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("Max Concentration;",InterferenceRange.max().round(2), 'mM')
print("Min Concentration:",InterferenceRange.min().round(4), 'mM')
print("Mean Concentration:",InterferenceRange.mean().round(2), 'mM')
```

```python
# Sort the samples to find the highest and lowest values
sorted_samples = np.sort(InterferenceRange)
# Get the 50 lowest values
lowest_50 = sorted_samples[:100]
# Get the 50 highest values
highest_50 = sorted_samples[-100:]
# Print the 50 lowest values
print("50 Lowest Concentrations:", np.round(lowest_50, 4))
# Print the 50 highest values
print("50 Highest Concentrations:", np.round(highest_50, 1))
```

## Mimic Tissue Range Distribution

```python
## Design a concentrations distribution for mimicking the distribution seen in experimental murine tissue spectra

Dist1 = (2/5)*np.random.lognormal(mean = 0, sigma = 1.5, size = 10000)
for i in np.arange(10000):
    if Dist1[i] < 0.005:
        Dist1[i] += np.random.uniform(0.005,0.01)
    if Dist1[i] > 20:
        Dist1[i] = np.random.uniform(0.005,1)
        
        
Dist2 = (1/5)*np.random.lognormal(mean = 0, sigma = 1.5, size = 10000)
for i in np.arange(10000):
    if Dist2[i] < 0.005:
        Dist2[i] += np.random.uniform(0.005,0.01)
    if Dist2[i] > 20:
        Dist2[i] = np.random.uniform(0.005,1)   
        

MimicTissueRange = np.concatenate((Dist1, Dist2)) 

# Plot the histogram of the bimodal distribution
plt.hist(MimicTissueRange, bins=100, density=False, alpha=0.6, color='g')
plt.title('Mimic Tissue Range Concentrations Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("Max Concentration;",MimicTissueRange.max().round(2), 'mM')
print("Min Concentration:",MimicTissueRange.min().round(4), 'mM')
print("Mean Concentration:",MimicTissueRange.mean().round(2), 'mM')


# Sort the samples to find the highest and lowest values
sorted_samples = np.sort(MimicTissueRange)
# Get the 50 lowest values
lowest_50 = sorted_samples[:50]
# Get the 50 highest values
highest_50 = sorted_samples[-50:]
# Print the 50 lowest values
print("50 Lowest Concentrations:", np.round(lowest_50, 4))
# Print the 50 highest values
print("50 Highest Concentrations:", np.round(highest_50, 1))
```

## High Dynamic Range Dataset Distribution

```python
# Practice generating a bimodal distribution that produces primarily lower concentration examples
Dist1 = 0.4*np.random.gamma(0.5, 0.6, size=10000)
Dist2 = 20 - np.random.gamma(1.75, 4, size=4000)
Dist3 = np.random.uniform(0 ,0.02, size=1000)
# Several loops for maintaining the target concentration range
for i in np.arange(10000):
    if Dist1[i] < 0:
        Dist1[i] = abs(Dist1[i])
for i in np.arange(4000):
    if Dist2[i] < 0:
        Dist2[i] = abs(Dist2[i])
for i in np.arange(10000):
    if Dist1[i] > 20:
        Dist1[i] = 20
for i in np.arange(4000):
    if Dist2[i] > 20:
        Dist2[i] = 20
HighDynRange2 = np.concatenate((Dist1, Dist2, Dist3))
for i in np.arange(15000):
    HighDynRange2[i] += np.random.choice([np.random.uniform(0.005,0.01),
                                          np.random.uniform(0.005,0.05),
                                          np.random.uniform(0.005,0.01)])

# Plot the histogram of the bimodal distribution
plt.hist(HighDynRange2, bins=100, density=True, alpha=0.6, color='g')
plt.title('High Dynamic Range Distribution 2')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()


print("Max Concentration;",HighDynRange2.max().round(2), 'mM')
print("Min Concentration:",HighDynRange2.min().round(4), 'mM')
print("Mean Concentration:",HighDynRange2.mean().round(2), 'mM')
```

```python
# Sort the samples to find the highest and lowest values
sorted_samples = np.sort(HighDynRange2)
# Get the 50 lowest values
lowest_50 = sorted_samples[:50]
# Get the 50 highest values
highest_50 = sorted_samples[-50:]
# Print the 50 lowest values
print("50 Lowest Concentrations:", np.round(lowest_50, 4))
# Print the 50 highest values
print("50 Highest Concentrations:", np.round(highest_50, 1))
```
